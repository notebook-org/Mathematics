```python
import numpy as np
from scipy import linalg as la
import matplotlib.pyplot as pt
```

```python
A = np.array([[-0.1005,1.0939,2.0428,4.4599],[-1.0880,-0.1444,5.9859,-3.0481],[-2.0510,-5.9709,-0.1387,1.9229],[-4.4575,3.0753,-1.8847,-0.1164]])
A
```

```python
w,v = la.eig(A)
```

```python
x0 = np.array([1,0,0,0])
x = x0.copy()
```

```python
n = 500
t = np.linspace(0,40,n)
```

```python
for i in range(1,n):
    x = np.column_stack((x,la.expm(t[i]* A) @ x0))
```

```python
pt.plot(t,x[0,:])
pt.show()
```

```python
pt.plot(t,x[1,:])
pt.show()
```

```python
pt.plot(t,x[2,:])
pt.show()
```

```python
pt.plot(t,x[3,:])
pt.show()
```
