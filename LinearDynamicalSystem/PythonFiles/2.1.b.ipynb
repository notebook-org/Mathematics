```python
import numpy as np
import matplotlib.pyplot as pt
```

Initializing data variables

```python
G = np.array([[1,0.2,0.1],[0.1,2,0.1],[0.3,0.1,3]])
gamma = 3
alpha = 1.2
sigma = 0.01
```

Initializing computation parameters

```python
t_max = 100
```

Initializing objective variables

```python
dim_G = np.shape(G)[0]
p = np.zeros( (dim_G, t_max) )
```

```python
A = np.zeros((dim_G, dim_G))
```

```python
for i in range(dim_G):
    for j in range(dim_G):
        if i != j:
            A[i][j] = alpha * gamma * (G[i][j]/G[i][i])
```

```python
print(A)
```

```python
b = np.zeros(dim_G)
```

```python
for i in range(dim_G):
    b[i] = (alpha * gamma * sigma) / G[i][i] 
```

```python
for i in range(1,t_max):
    p[:,i] = np.dot(A, p[:,i-1]) + b
```

```python
pt.plot(range(t_max), p[0,:],'r')
pt.plot(range(t_max), p[1,:],'g')
pt.plot(range(t_max), p[2,:],'b')
pt.show()
```

```python
B = np.zeros((dim_G,dim_G))
for i in range(dim_G):
    B[i][i] = G[i][i]
s = np.dot(B,p)
```

```python
pt.plot(range(t_max), s[0,:],'r')
pt.plot(range(t_max), s[1,:],'g')
pt.plot(range(t_max), s[2,:],'b')
pt.show()
```

```python
C = G
for i in range(dim_G):
    B[i][i] = 0
c = sigma * np.ones((dim_G, t_max))
q = np.dot(C,p) + c
```

```python
pt.plot(range(t_max), q[0,:],'r')
pt.plot(range(t_max), q[1,:],'g')
pt.plot(range(t_max), q[2,:],'b')
pt.show()
```

```python
SINR = s/q
```

```python
pt.plot(range(t_max), SINR[0,:],'r')
pt.plot(range(t_max), SINR[1,:],'g')
pt.plot(range(t_max), SINR[2,:],'b')
pt.show()
```
